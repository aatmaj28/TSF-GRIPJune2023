## Conclusions and Inferences
**On performing the EDA we understand the following**

- Attacks on private citizens and property came in first, then on the military, police, and other governmental entities.
- The Middle East and North Africa were the most severely hit regions. Most residents of this area were either killed or seriously injured.
- Iraq was the nation that saw the greatest number of terrorist attacks and fatalities as well as injuries.
- Baghdad was the State and City most severely impacted.
- The most terror incidents occurred in 2014 during the previous ten years. This year had approximately 16500 attacks, or an average of 45 attacks per day.
- Explosion/bombing attacks were most frequently used.
- Terrorists' go-to weapon of choice has consistently been explosives.
- Since 2012, the Taliban have increased their activity and are the main perpetrators of terrorist assaults.



**Suggestions from the Analysis**
- Stronger security and surveillance should be offered, especially in areas with dense populations because private citizens and property are frequently targeted.
- Particularly in the Middle East and North African Regions, more surveillance is needed.
- Strict border controls should be put in place to stop the transfer of explosives between regions.
